```python
# Identify categorical columns
categorical_col =df.select_dtypes(include ='object').columns.tolist()
categorical_col.remove('income')
```

```python
# Determine number of unique values in each categorical column
unique_counts =df[categorical_col].nunique()
unique_counts
```

```python
# Apply One-Hot Encoding for categorical variables with < 5 unique values
one_hot_cols =unique_counts[unique_counts < 5].index.tolist()
df_one_hot_encoded =pd.get_dummies(df[one_hot_cols], drop_first =True)
df_one_hot_encoded
```

```python
# Apply Label Encoding for categorical variables with >= 5 unique values
label_cols =unique_counts[unique_counts > 5].index.tolist()
df_label_encoded =df[label_cols].apply(LabelEncoder().fit_transform)
df_label_encoded
```

```python
# Combine all encoded categorical features
df_encoded =pd.concat([df.drop(columns =one_hot_cols + label_cols + ['income']),
                       df_one_hot_encoded, df_label_encoded], axis =1)
```

### 3. Feature Engineering:

```python
import numpy as np

# Copy the encoded dataframe to preserve original
df_fe = df_encoded.copy()

# 1. Create new feature: capital_diff = capital_gain - capital_loss
df_fe['capital_diff'] = df_fe['capital_gain'] - df_fe['capital_loss']

# 2. Create new feature: age_group (categorical bucketing of age)
df_fe['age_group'] = pd.cut(df_fe['age'], bins=[0, 25, 45, 65, 100], labels=['Young', 'Adult', 'Senior', 'Elder'])

# Convert age_group to one-hot encoding (as it's categorical)
df_fe = pd.concat([df_fe, pd.get_dummies(df_fe['age_group'], prefix='age_group', drop_first=True)], axis=1)
df_fe.drop(columns='age_group', inplace=True)

# Check skewness of numerical features
skewness = df_fe[['capital_gain', 'capital_loss', 'fnlwgt']].skew()

# Apply log(1 + x) transformation to capital_gain (highly skewed)
df_fe['capital_gain_log'] = np.log1p(df_fe['capital_gain'])

# Show skewness values and transformed data preview
skewness, df_fe[['capital_gain', 'capital_gain_log']].head()
```

### 4. Feature Selection:

```python
'''
Outlier Detection and Removal
Method Used: Isolation Forest (contamination=0.01)

Outliers Removed: ~1% of the data (326 records)

Remaining Records: 32,235

 Why Remove Outliers?
Outliers can distort model training by:

Skewing mean-based metrics and distributions

Reducing accuracy in linear models and distance-based algorithms

Increasing overfitting in models that try to capture noise
'''
```

```python
from sklearn.ensemble import IsolationForest

# Apply Isolation Forest to numerical features only
numerical_features = df_fe.select_dtypes(include=[np.number])
iso_forest = IsolationForest(contamination=0.01, random_state=42)
outliers = iso_forest.fit_predict(numerical_features)

# Add outlier flag and filter out outliers
df_fe['outlier'] = outliers
df_clean = df_fe[df_fe['outlier'] == 1].drop(columns='outlier')

# Calculate percentage of data removed
removed_percentage = 100 * (len(df_fe) - len(df_clean)) / len(df_fe)

# Show shape after outlier removal and % removed
df_clean.shape, removed_percentage
```
