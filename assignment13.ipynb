## Interview Questions:

```python
#1. What are some common hyperparameters of decision tree models, and how do they affect the model's performance?
#Decision tree models (like those in scikit-learn or XGBoost) have several hyperparameters that control their complexity, accuracy, and generalization. Here's a summary of the most important ones:
print('''
1. max_depth
    What it does: Limits the maximum depth of the tree.

    Effect:

        Low value â†’ underfitting (tree too shallow).

        High value â†’ overfitting (tree too deep).
2. min_samples_split
    What it does: Minimum number of samples required to split an internal node.

    Effect:

        Higher value â†’ more conservative splits â†’ less overfitting.

        Lower value â†’ more splits â†’ risk of overfitting.
3. min_samples_leaf
    What it does: Minimum number of samples required to be at a leaf node.

    Effect:

        Helps smooth the model by requiring larger leaf nodes.

        Higher value â†’ more generalization, less overfitting.
4. max_features
    What it does: Number of features to consider when looking for the best split.

    Effect:

        Smaller value introduces randomness â†’ helps in ensemble methods (like Random Forest).

        Larger value â†’ more accurate splits, but risk of overfitting.

5. max_leaf_nodes
    What it does: Limits the number of leaf nodes.

    Effect:

        Controls model complexity.

        Prevents overfitting by stopping further growth once the limit is reached.

6. criterion
    What it does: The function used to measure the quality of a split.

    gini: Gini impurity (default).

    entropy: Information gain.

    Effect:

        Slight differences in performance; gini is faster, entropy may be more informative.

7. splitter
    What it does: Strategy used to choose the split at each node (best or random).

    Effect:

        best: Finds optimal split (can overfit).

        random: Adds randomness (useful in ensemble settings).
''')
```
